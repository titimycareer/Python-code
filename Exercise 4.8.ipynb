# Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Step 1

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(10)
```

```python
# Step 2

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# loyalty_flag frequency

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Step 3

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(10)
```

#  The team now wants to target different types of spenders in their marketing campaigns.

```python
# Step 1

ords_prods_merge['Ave_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(1)
```

```python
# Step 2

ords_prods_merge.loc[ords_prods_merge['Ave_price'] >=10, 'Spender_pattern'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['Ave_price'] < 10, 'Spender_pattern'] = 'Low spender'
```

```python
# Spender_pattern frequency

ords_prods_merge['Spender_pattern'].value_counts(dropna = False)
```

There are more Low spenders than High spender.

#  Determine frequent versus non-frequent customers

```python
# Step 1

ords_prods_merge['median_regularity'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge['days_since_prior_order'].head(10)
```

```python
# Step 2

ords_prods_merge.loc[ords_prods_merge['median_regularity'] > 20, 'user_regularity'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_regularity'] <= 20) & (ords_prods_merge['median_regularity'] > 10), 'user_regularity'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_regularity'] <= 10, 'user_regularity'] = 'Frequent customer'
```

```python
# user_regularity frequency

ords_prods_merge['user_regularity'].value_counts(dropna = False)
```

There are more Frequent customer, followed by Regular customer and lastly by Non-frequent customer.

```python
# Step 3

ords_prods_merge[['user_id', 'user_regularity', 'days_since_prior_order', 'median_regularity']].head(10)
```
